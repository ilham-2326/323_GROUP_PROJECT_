# **FEATURE ENGINEERING**

## Deriving rental_yield

````python
# 1.  Then add your derived features
# Derived feature: rental_yield = annual office rent / total sale value (a proper ratio).
# avg_sale_value is the per-district mean of actual_worth on sales (from the transactions step).
# NOTE: divide by avg_sale_value, NOT avg_sale_price (price per m^2) -> that gave the ~488% bug.
final_df["rental_yield"] = final_df["avg_rent"] / final_df["avg_sale_value"]

# IQR-cap to remove proxy extremes (same outlier method used on the other features)
q1, q3 = final_df["rental_yield"].quantile([0.25, 0.75])
iqr = q3 - q1
lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
n_capped = ((final_df["rental_yield"] < lower) | (final_df["rental_yield"] > upper)).sum()
final_df["rental_yield"] = final_df["rental_yield"].clip(lower, upper)
print(f"rental_yield = avg_rent / avg_sale_value; IQR-capped to [{lower:.4f}, {upper:.4f}] for {n_capped} district(s)")
````

## Build survey weights from `normalized`

````python
import pandas as pd

# Re-define survey and normalized within this cell for standalone execution
df = pd.read_excel("survey_responses.xlsx")
df.to_csv("survey.csv", index=False)
survey = pd.read_csv("survey.csv")
scores = survey.drop(columns=['Timestamp']).mean()  # mean over all survey responses
normalized = scores / scores.sum()

# Map our 8 data columns to the correct survey feature weights
# using the 'normalized' Series already computed from the survey CSV
col_to_feature = {
    'avg_sale_price'   : 'Feature 1 : Capital Appreciation Potential ',
    'transaction_count': 'Feature 3 : Commercial Liquidity & Deal Volume ',
    'parking_score'    : 'Feature 4 : Building Parking Sufficiency ',
    'rental_yield'     : 'Feature 5 : Current Rental Yield Baseline ',
    'contract_count'   : 'Feature 6 : Corporate Space Absorption Velocity ',
    'avg_rent'         : 'Feature 7 : District Supply Tightness Rate ',
    'metro_score'      : 'Feature 8 : Transit-Oriented Accessibility Index ',
    'mall_score'       : 'Feature 9 : Business Lifestyle Convenience Index ',
}

# Pull the raw weights for our 8 columns
raw_weights = {col: normalized[feat] for col, feat in col_to_feature.items()}

# Renormalise so they sum to 1.0 (absorbs the missing features 2, 10, 11, 12)
total = sum(raw_weights.values())
feature_weights = {col: w / total for col, w in raw_weights.items()}

print('Adjusted weights (sum = 1.0):')
for col, w in feature_weights.items():
    print(f'  {col:<22}  {w:.6f}')
print(f"  {'TOTAL':<22}  {sum(feature_weights.values()):.6f}")
````

## **WEIGHTED COMPOSITE SCORE**

Raw columns are on completely different scales — `avg_sale_price` is in AED,
`transaction_count` can exceed 10,000, while `mall_score` is 0–5.
Without normalisation, large-magnitude columns dominate the score regardless of weight.
Min-Max scaling brings every feature to [0, 1] before applying weights.

````python
from sklearn.preprocessing import MinMaxScaler

# Correct the feature_cols list to include only existing columns
feature_cols = [
    "avg_sale_price",
    "transaction_count",
    "parking_score",
    "rental_yield",
    "avg_rent",
    "contract_count",
    "metro_score",
    "mall_score",
]

# Min-Max normalise all feature columns
scaler = MinMaxScaler()
norm_col_names = [col + '_norm' for col in feature_cols]
final_df[norm_col_names] = scaler.fit_transform(final_df[feature_cols])

print('\nNormalised features — first 3 rows:')
display(final_df[['area_id'] + norm_col_names].head(3))
````

## Compute weighted model
For each area:
```
hotspot_score = Σ (normalised_feature_i × weight_i)
```

````python
# Weighted sum of normalised features
final_df['hotspot_score_raw'] = sum(
    final_df[col + '_norm'] * feature_weights[col]
    for col in feature_cols
)

# Scale to 0–100
score_min = final_df['hotspot_score_raw'].min()
score_max = final_df['hotspot_score_raw'].max()
final_df['hotspot_score'] = (
    (final_df['hotspot_score_raw'] - score_min) / (score_max - score_min)
) * 100

# Rank (1 = best hotspot)
final_df['rank'] = final_df['hotspot_score'].rank(ascending=False).astype(int)

result_df = final_df[['area_id', 'hotspot_score', 'rank']].sort_values('rank').reset_index(drop=True)

print('=== TOP 15 HOTSPOT AREAS ===')
print(result_df.head(15).to_string(index=False))
````
